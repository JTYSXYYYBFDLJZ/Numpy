```python
#63.创建一个具有name属性的数组类
class NameArray(np.ndarray):
    def __new__(cls, array, name="no name"):
        #　-类.__new__方法，创建一个对象
        #__new__() 方法是在类准备将自身实例化时调用，静态方法
        obj = np.asarray(array).view(cls)
        obj.name = name
        return obj
    def __array_finalize__(self, obj):
        if obj is None: return
        self.info = getattr(obj, 'name', "no name")
 
Z = NameArray(np.arange(10), "range_10")
print (Z.name)
```

```python
#75. 如何通过滑动窗口计算一个数组的平均数?
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)#计算轴向元素累加和，返回由中间结果组成的数组
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
Z = np.arange(20)
print(moving_average(Z, n=3))
```

```python
#76.思考以为数组Z，构建一个二维数组，其第一行是(Z[0],Z[1],Z[2])， 然后每一行移动一位，最后一行为 (Z[-3],Z[-2],Z[-1]) 

from numpy.lib import stride_tricks
def rolling(a, window):
    shape = (a.size - window + 1, window)
    strides = (a.itemsize, a.itemsize)
    return stride_tricks.as_strided(a, shape=shape, strides=strides)#这个函数可以高效地切分一个数组为不同的shape块，strides跨过多少比特
Z = rolling(np.arange(10), 3)
print(Z)
```

```python
#78.思考两组点集P0和P1去描述一组线(二维)和一个点p,如何计算点p到每一条线 i (P0[i],P1[i])的距离？
def distance(P0, P1, p):
    T = P1 - P0
    L = (T**2).sum(axis=1)#直线长度
    U = -((P0[:,0]-p[...,0])*T[:,0] + (P0[:,1]-p[...,1])*T[:,1]) / L
    U = U.reshape(len(U),1)
    D = P0 + U*T - p
    return np.sqrt((D**2).sum(axis=1))
P0 = np.random.uniform(-10,10,(10,2))
print(P0)
P1 = np.random.uniform(-10,10,(10,2))
print(P1)
p  = np.random.uniform(-10,10,( 1,2))
print(p)
print(distance(P0, P1, p))
```

```python
#85.创建一个满足 Z[i,j] == Z[j,i]的二维数组子类
class Symetric(np.ndarray):
    def __setitem__(self, index, value):
        i,j = index
        super(Symetric, self).__setitem__((i,j), value)
        super(Symetric, self).__setitem__((j,i), value)
def symetric(Z):
    return np.asarray(Z + Z.T - np.diag(Z.diagonal())).view(Symetric)
S = symetric(np.random.randint(0,10,(5,5)))
S[2,3] = 42
print(S)
```
